```python
import google.generativeai as genai
import re
import pandas as pd
import numpy as np

genai.configure(api_key="AIza ")
model2 = genai.GenerativeModel("gemini-1.5-flash")

def get_gemini_response(prompt):
    response = model2.generate_content(prompt)
    return response.text

def sub_part_a_query(query):
    query = query.replace('"', "'")
    out = get_gemini_response(f"The following sentence describes a task. Try and split the task into smaller sub tasks. Sentence: {query}. Do not make any extra assumptions that are not in the sentence, retain all the verbs and keywords. Your response should be a sequence of tasks, in the form of a python list, and each subquery must only have 1 action. The order of subqueries should be logical, and need not copy the order of words in the query.")
    list_as_string = re.search(r"\[(.|\n)+\]", out).group()
    tasks = [item.strip() for item in list_as_string[1:-1].replace('"', '').split(',')]
    print(tasks)
    return tasks

def extract_argument_values(query, relevant_arguments):
    extracted_args = {}
    for _, row in relevant_arguments.iterrows():
        arg_name = row["Argument Name"]
        if arg_name:
            if "array of strings" in row["Argument Description"]:
                values = [val.strip("'") for val in query.split() if val.startswith("'") and val.endswith("'")]
                if values:
                    extracted_args[arg_name] = values
            elif "boolean" in row["Argument Description"]:
                if "need response" in query:
                    extracted_args[arg_name] = True
            elif "string" in row["Argument Type"]:
                words = query.split()
                for word in words:
                    if word.startswith("'") and word.endswith("'"):
                        extracted_args[arg_name] = word.strip("'")
    return extracted_args

def keyword_match_score(query, df):
    """Calculate a score based on exact keyword matches."""
    score_dict = {}
    keywords = set(query.split())

    for index, row in df.iterrows():
        tool_description_keywords = set(row['Description'].split())
        match_count = len(keywords.intersection(tool_description_keywords))
        score_dict[index] = match_count

    return score_dict

def merge_duplicate_tools_in_json(input_j):
    new_j = []
    last_tool = {'tool_name': None}
    for lst in input_j:
        tool = lst[0]
        if tool['tool_name'] == last_tool['tool_name']:
            last_tool['arguments'] += tool['arguments']
        else:
            new_j.append(tool)
        last_tool = tool
    return [[tool] for tool in new_j]

def find_best_tool_and_arguments(query, df):
    # Calculate keyword match scores
    keyword_scores = keyword_match_score(query, df)

    # Compute cosine similarities
    query_embedding = model.encode(query.lower())
    tool_similarities = util.pytorch_cos_sim(query_embedding, model.encode(df['Description'].tolist()))

    # Combine scores into a weighted score
    combined_scores = {}

    for index in range(len(df)):
        # Access the similarity score for the corresponding tool
        similarity_score = tool_similarities[0][index].item()  # Accessing first row for query similarity

        combined_scores[index] = (keyword_scores.get(index, 0)*1) + similarity_score  # CHANGE WEIGHT HERE

    best_tool_index = max(combined_scores, key=combined_scores.get)

    tool_name = df.iloc[best_tool_index]['Tool']

    # Compute similarities with argument+ tool descriptions
    df['Combined Description'] = df['Argument Description'].fillna('') + ' ' + df['Description'].fillna('')
    argument_similarities = util.pytorch_cos_sim(query_embedding, model.encode(df['Combined Description'].fillna('').tolist()))

    for index in range(len(df)):
        # Access the similarity score for the corresponding tool
        similarity_score = argument_similarities[0][index].item()  # Accessing first row for query similarity

        combined_scores[index] = (keyword_scores.get(index, 0)*1) + similarity_score  # CHANGE WEIGHT HERE

    relevant_arguments = df.iloc[argument_similarities.argsort(descending=True).squeeze()[:1]]
#     relevant_arguments = max(combined_scores, key=combined_scores.get)

    extracted_arguments = extract_argument_values(query, relevant_arguments)

    tool_name_row = df[df['Argument Name'] == relevant_arguments['Argument Name'].values[0]]['Tool']

    if not tool_name_row.empty:
        tool_name = tool_name_row.values[0]

    return tool_name, extracted_arguments

def generate_json(tool_name, extracted_arguments):
    arguments_list = [{"argument_name": arg_name, "argument_value": arg_value} for arg_name, arg_value in extracted_arguments.items()]
    return [{"tool_name": tool_name, "arguments": arguments_list}]

def process_query(query, df):
    j=[]
    qs= sub_part_a_query(query.lower())

    for q in qs:
        tool_name, extracted_arguments = find_best_tool_and_arguments(q, df)
        json_output = generate_json(tool_name, extracted_arguments)
        j.append(json_output)
    j= merge_duplicate_tools_in_json(j)
    return j

import json
def pretty_print_json(json_data):
    if isinstance(json_data, pd.Series):
        json_data = json_data.to_list()

    elif isinstance(json_data, pd.DataFrame):
        json_data = json_data.to_dict(orient='records')

    return json.dumps(json_data, indent=4)

from sklearn.cluster import KMeans

from sentence_transformers import SentenceTransformer

# Load a pre-trained model for generating embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to generate hypotheses from the initial query
def generate_hypotheses(initial_query):
    # Simple example: generate hypotheses by varying the query slightly
    return [f"{initial_query} variation {i}" for i in range(5)]

# Function to conduct cluster analysis on the generated hypotheses
def cluster_hypotheses(hypotheses):
    embeddings = model.encode(hypotheses)
    kmeans = KMeans(n_clusters=3)  # Number of clusters can be adjusted
    kmeans.fit(embeddings)
    clusters = kmeans.labels_
    return clusters, embeddings

# Function to refine sub-queries from clusters
def refine_sub_queries(hypotheses, clusters):
    refined_queries = []
    for cluster_id in np.unique(clusters):
        cluster_hypotheses = [hyp for hyp, cl in zip(hypotheses, clusters) if cl == cluster_id]
        # Select a representative hypothesis from each cluster (e.g., the first one)
        refined_queries.append(cluster_hypotheses[0])
    return refined_queries

# Function to iteratively enhance performance
def autoregressive_planning(initial_query, max_iterations=5):
    current_queries = [initial_query]
    iteration = 0

    while iteration < max_iterations:
        print(f"Iteration {iteration + 1}:")
        
        # Generate hypotheses based on current queries
        hypotheses = []
        for query in current_queries:
            hypotheses.extend(generate_hypotheses(query))

        # Conduct cluster analysis
        clusters, embeddings = cluster_hypotheses(hypotheses)

        # Refine queries from clusters
        current_queries = refine_sub_queries(hypotheses, clusters)

        print(f"Refined Queries: {current_queries}")
        
        # Check if the query requirements are met (placeholder for actual check)
        if len(current_queries) == 1 and "satisfied" in current_queries[0]:
            print("Query requirements satisfied.")
            break
        
        iteration += 1

    return current_queries
```

```python
with open('/kaggle/input/sfadvadv/few_shots.txt', 'r') as file:
    few_shot_examples = file.read()

def optimize_with_llm(output_json):
    # Convert the output JSON to text for LLM processing
    json_text = pretty_print_json(output_json)

    # Pass the output to Gemini model with stricter instructions
    prompt = f"""
    Here are some example queries and their solutions:\n{few_shot_examples}
    
    Now, optimize the following tool selection and arguments using this format:\n{json_text}
    
    Please ensure that:
    1. You do not add extra tools or arguments beyond what is necessary.
    2. Only use tools and arguments directly related to the query.
    3. Do not include any irrelevant tools or arguments unless they are critical for answering the query.
    """
    
    # Get the response from the Gemini LLM model
    optimized_response = get_gemini_response(prompt)
    
    # Print the raw LLM response for debugging
    print("LLM Response:\n", optimized_response)
    
    try:
        # Split the response to isolate JSON part (handle extra text or comments)
        json_parts = re.findall(r"\{(.|\n)*\}", optimized_response)
        if json_parts:
            json_part = json_parts[0]  # Take only the first valid JSON part
            optimized_output = json.loads(json_part)
        else:
            raise ValueError("LLM response does not contain valid JSON.")
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        raise
    
    return optimized_output

        

def process_query(query, df):
    j = []
    qs = sub_part_a_query(query.lower())

    for q in qs:
        tool_name, extracted_arguments = find_best_tool_and_arguments(q, df)
        json_output = generate_json(tool_name, extracted_arguments)
        j.append(json_output)
    
    j = merge_duplicate_tools_in_json(j)
    
    # Optimize the generated JSON by passing it through the LLM
    optimized_json = optimize_with_llm(j)
    
    return optimized_json

# Your existing code remains unchanged
```
